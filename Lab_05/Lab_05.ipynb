```python
from datascience import *
import numpy as np
path_data = 'PopEst_IN_counties.csv'
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

1) Read PopEst_IN_counties.csv in as a table.

2) Make a table with the Rural_)Urban column for 2013.

3) Make a table of counts for each Ruran-Urban code for 2013.

4) Sort this table from highest to lowest counts.

5) Make a bar chart based on the table.  Which Codes are the two most common?

```python
# 1) Read PopEst_IN_counties.csv in as a table.
PopEst_IN = Table().read_table(path_data)
PopEst_IN
```

```python
# 2) Make a table with the Rural_)Urban column for 2013.
Rural_Urban2013 = PopEst_IN.select("Rural-urban_Continuum Code_2013")
Rural_Urban2013
```

```python
# 3) Make a table of counts for each Ruran-Urban code for 2013.
Rural_Urban_count = Rural_Urban2013.group("Rural-urban_Continuum Code_2013")
Rural_Urban_count
```

```python
# 4) Sort this table from highest to lowest counts.
new = Rural_Urban_count.sort('count', descending = True)
new
```

```python
# 5) Make a bar chart based on the table. Which Codes are the two most common?
new.barh('Rural-urban_Continuum Code_2013')
```
